```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from multiprocessing.dummy import Pool
%matplotlib inline

import matplotlib as mpl
mpl.rcParams['axes.labelsize'] = 20
mpl.rcParams['xtick.labelsize'] = 14
mpl.rcParams['ytick.labelsize'] = 14
mpl.rcParams['axes.titlesize'] = 20
mpl.rcParams['legend.fontsize'] = 16
```

```python
def extract_features(df):
    ## For example:
    df['day_of_week'] = df['date'] % 7
```

```python
def cleave_test(df, weeks=1):
    # Cleave off last weeks
    last_date = df['date'].max()
    df = df[df['date'] <= last_date - 7*(weeks-1)]

    last_date -= 7*(weeks-1)
    df_train = df[df['date'] <= last_date - 7].copy()
    df_train['date'] -= last_date

    df_test = df[df['date'] > last_date - 7]
    df_test = df_test.groupby('id', as_index=False).head(1).copy()
    targets = np.zeros(len(np.unique(df['id'])))
    targets[df_test['id'].values-1] = df_test['sum'].values
    df_test['target'] = df_test['sum']
    return df_train, targets
```

```python
def predict(df_train, seed=0):
    np.random.seed(seed)
    predictions = df_train.groupby('id').apply(lambda x: np.random.randint(0, 17)).values
    return predictions
```

```python
def cross_validate(cv=5, **hyperparameters):
    accuracies = []
    for weeks in range(1, cv):
        df_train, df_test = cleave_test(df, weeks=weeks)
        predictions = predict(df_train, **hyperparameters)
        accuracy = (predictions == df_test).mean()
        accuracies.append(accuracy)
    return np.mean(accuracies), accuracies
```

```python
def linear_search(parameter_name, extent, **hyperparameters):
    score = []
    for i in extent:
        hyperparameters[parameter_name] = i
        score.append(cross_validate(**hyperparameters)[0])
    plt.plot(extent, score)
    plt.xlabel(parameter_name)
    plt.ylabel('Accuracy')
    plt.title('Optimal %s: %0.4f'%(parameter_name, extent[np.argmax(score)]))
```

```python
def select_parameters(parameters):
    random_parameters = dict()
    for k, v in parameters.items():
        random_parameters[k] = v[np.random.randint(0, len(v))]
    return random_parameters

def random_search(parameters, runs=1000, jobs=-1):
    if jobs < 0:
        map_ = map
    else:
        map_ = Pool(jobs).map
    random_parameters = [select_parameters(parameters) for _ in range(runs)]
    results = map_(lambda p: cross_validate(**p)[0], random_parameters)
    return random_parameters[np.argmax(results)], np.max(results)
```

```python
def submit(name, cv=False, **hyperparameters):
    df = pd.read_csv('data/all.csv').sort_values(by=['id', 'date'])
    extract_features(df)
    if cv:
        print cross_validate()[0]
    df_train, _ = cleave_test(df, 0)
    prediction = predict(df_train, **hyperparameters)
    submission = pd.DataFrame(prediction, columns=['sum']).reset_index()
    submission['id'] = submission['index']+1
    path = '{}.csv'.format(name)
    submission[['id', 'sum']].to_csv(path, index=None)
    print 'Stored submission in {}'.format(path)
```
