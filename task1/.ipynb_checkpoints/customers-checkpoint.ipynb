```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import lightgbm
%matplotlib inline
plt.style.use('ggplot')
```

## calendar

```python
data = pd.read_csv('sample/data.csv')[['id', 'date', 'sum']]
days_count = data['date'].max()
persons_count = len(data.groupby('id'))
```

```python
ends = data[['id', 'date']].groupby('id').max()['date']
ends.plot.hist(bins=50)
plt.xlabel('day')
plt.ylabel('count')
plt.savefig('images/ends.pdf')
```

```python
former_idxs = np.where(ends.values < 400)[0]
np.mean((ends.values < 400)) 
```

```python
current_idxs = set(range(persons_count)) - set(former_idxs)
current_idxs = np.array(list(current_idxs))
```

```python
starts = data[['id', 'date']].groupby('id').min()['date']
starts.plot.hist(bins=50)
plt.xlabel('day')
plt.ylabel('count')
plt.savefig('images/starts.pdf')
```

```python
points = data.groupby('id').count()['date'].as_matrix()
points = np.cumsum(points)
points = np.hstack((0, points))
```

```python
calendar = np.zeros((persons_count, days_count), int)
for j in range(persons_count):
    calendar[j][data[points[j]:points[j + 1]]['date'] - 1] = data[points[j]:points[j + 1]]['sum']
```

```python
calendar = calendar[current_idxs, :]
```

```python
plt.bar(np.arange(17), np.log(np.bincount(calendar.ravel())))
plt.xlabel('labels')
plt.ylabel('log count')
plt.xticks(np.arange(17))
plt.savefig('images/label_distr_1.pdf')
```

```python
calendar[calendar > 8] = 8
plt.bar(np.arange(9), np.log(np.bincount(calendar.ravel())))
plt.xlabel('labels')
plt.ylabel('log count')
plt.xticks(np.arange(9))
plt.savefig('images/label_distr_2.pdf')
```

```python
def make_labels(week):
    idxs = np.sum(week, axis=1) != 0
    nonzero_labels = week[idxs][np.cumsum(
        np.cumsum(week[idxs] > 0, axis=1), axis=1) == 1]
    labels = np.zeros(week.shape[0], int)
    labels[idxs] = nonzero_labels
    return labels
```

```python
train_labels = make_labels(calendar[:, -7:])
train_data = calendar[:, :-7]
test_data = calendar[:, 7:]
```

```python
np.save('sample/train_labels.npy', train_labels)
np.save('sample/train_data.npy', train_data)
np.save('sample/test_data.npy', test_data)
```

# lightgbm

```python
model = lightgbm.LGBMClassifier(n_estimators=400, learning_rate=0.01, n_jobs=-1, max_depth=6)
```

```python
def make_features_weighted(data):
    weights = np.arange(data.shape[1], dtype=float)
    weights /= np.sum(weights)
    
    counters = pd.concat(
        [((data == j)*weights).sum(axis=1) for j in range(9)], axis=1)
    
    data = data*weights
    return pd.concat((
        data[data!=0].mean(axis=1),
        data.mean(axis=1),
        data[data!=0].std(axis=1),
        data.std(axis=1),
        counters),
        axis=1)
```

```python
%%time
model.fit(make_features_weighted(pd.DataFrame(train_data)), train_labels)
```

```python
model.feature_importances_
```

```python
test_labels = model.predict(make_features_weighted(pd.DataFrame(test_data)))
```

```python
np.save('sample/test_labels_lgbm.npy', test_labels)
```
